```python
Telegram_reviews_list=list(Telegram_reviews)

Telegram_reviews_list
```

```python
username = []
review=[]
user_score = []
thumbsup = []
date=[]

for x in range(0,20000):
    
    user_name=Telegram_reviews_list[0][x]["userName"]
    
    username.append(user_name)
    
    review_=Telegram_reviews_list[0][x]["content"]
    
    review.append(review_)
    
    score_=Telegram_reviews_list[0][x]["score"]
    
    user_score.append(score_)
    
    thumbsup_=Telegram_reviews_list[0][x]["thumbsUpCount"]
    
    thumbsup.append(thumbsup_)
    
    date_at=Telegram_reviews_list[0][x]["at"]
    
    date.append(date_at)

```

```python
print(len(username))
print(len(review))
print(len(user_score))
print(len(thumbsup))
print(len(date))
```

```python
import pandas as pd
```

```python
Telegram =pd.DataFrame({"Username":username, "Reviews": review,"Score":user_score , "ThumbsupCount" : thumbsup, "Date&Time":date})

Telegram
```

```python
# Let's pick only the reviews.text column.

review_df=Telegram[['Reviews']]
review_df
```

```python
review_df['Reviews']=review_df['Reviews'].astype(str)
```

```python
review_df.isnull().sum()
```

```python
review_df.dropna()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost
```

```python
# Create an Sentiment Intensity Analyzer object.

senti=SentimentIntensityAnalyzer()
```

```python
# STEP 1: Create a user defined lambda function.

fun=lambda x: senti.polarity_scores(x)['compound']
```

```python
# STEP 2: Apply the function on your data frame.

review_df['Compound']=review_df['Reviews'].apply(fun)

review_df
```

```python
# Let's create category labels as per the above slabs.

import numpy as np

review_df['Category']=pd.cut(review_df['Compound'],bins=[-1,-0.5,-0.2,0.2,0.5,1],labels=['High Neg','Neg','Neu','Pos','High Pos'])

review_df
```

```python
file_name = 'Compound Scores Dataset.xlsx'
review_df.to_excel(file_name)
```

```python
review_df.value_counts
```

## Building Machine Learning Classification Models based on our Compound Scores.

```python
# STEP 1: Define x and y.

x=review_df['Reviews']
y=review_df['Category']

# STEP 2: Splitting the data into train and test data.

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)

len(x_train), len(x_test), len(y_train), len(y_test)
```

To build a Machine Learning Model, we need to have x as a numerical column. In this case, x is purely a text data and thus we must convert this text into numbers by creating a DTM. You may use a Count DTM or a TF-IDF DTM.

In this case, we will use a TF-IDF DTM as it gives good results.

Let's build a DTM on x_train before building the model.

```python
# let's create a TF-IDF vectorizer.

from sklearn.feature_extraction.text import TfidfVectorizer

vec=TfidfVectorizer()
```

```python
# Let's create the DTM on x_train.

x_train_dtm=vec.fit_transform(x_train)
x_train_dtm
```

```python
y_train.value_counts()
```

```python
# Since data is imbalanced we will use SMOTE to balance the data. 

from imblearn.over_sampling import SMOTE

sm=SMOTE(random_state=42)
x_train_dtm_new,y_train_new=sm.fit_resample(x_train_dtm,y_train)
```

```python
y_train_new.value_counts()
```

Data is balanced to run predictions. 

## Building Decision Tree Model.

```python
# STEP 1: Create an object.

tree=DecisionTreeClassifier()

# STEP 2: Fit the model object on x and y.

model_dt=tree.fit(x_train_dtm_new,y_train_new)

model_dt
```

```python
x_test
```

```python
# Let's create the x_test_dtm.

x_test_dtm=vec.transform(x_test)
x_test_dtm
```

```python
# Let's predict the category on x_test_dtm.

y_test_pred_dt=model_dt.predict(x_test_dtm)

y_test_pred_dt
```

```python
# Let's find the accuracy of the model by creating a confusion matrix between y_test and y_test_pred.

pd.crosstab(y_test,y_test_pred_dt)
```

```python
y_test.shape
```

```python
# The accuracy of model matching both rows and column scores.

Acc_dt=(88+59+897+1150+997)/4000

Acc_dt
```

Accuracy of Decision Tree Model = 79.78%.
